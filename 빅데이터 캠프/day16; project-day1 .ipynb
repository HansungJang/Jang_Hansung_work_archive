```python
# Python3 샘플 코드 #
# '한동대' 키워드를 활용하여 22.08.01~ 23.07.01 분실 등록된 자료 불러오기
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests

url = 'http://apis.data.go.kr/1320000/LostGoodsInfoInqireService/getLostGoodsInfoAccTpNmCstdyPlace'
params ={'serviceKey' : '+y8zxHxSx9sqT6G4i9gNFkZaqH23k+h51QArnY82ML8D+J3CV1ke+ZAPvIlhtgKUAJPUFi6FaNWgcmehbiwzIA==', 'START_YMD' : '20220801', 'END_YMD' : '20230701', 'numOfRows' : '100', 'LST_PLACE':'한동대' , 'pageNo' : '1'}

page = requests.get(url, params=params)
print(page.content)
```

```python
soup =  BeautifulSoup(page.content,'html.parser')
print(soup.prettify())
```

```python
#3가지 정보를 추출 
# 1. lstplace  :분실장소, 2. lstymd: 분실기간, 3.prdtclnm : 분실 물품 종류

place = soup.find_all('lstplace')
pl_lst = [x.text for x in place] #분실장소

time = soup.find_all('lstymd')
tm_lst = [x.text for x in time] #분실시간

stuff = soup.find_all('prdtclnm')
sf_lst = [x.text for x in stuff] #분실물간

```

```python
# 인코딩이 필요; 같은 데이터에 대해서 제각기 표현하는 방법이 다름으로 통일이 필요 
filter_df = df.copy() #  본 데이터를 복사 
filter_df['분실 장소']
```

```python
# 데이터 가공; (2022~2023 데이터 만 추출;)

new_df = filter_df[:10]
new_df
```

```python
#다음 '분실 장소' 데이터 중복되는 표현 통일 

#중복되는 값에 대한 데이터 정렬
new_df.groupby('분실 장소').count()
```

```python
#데이터 통일 
#>> 한동대학교 / 버스 2가지 장소로 통일
for i in range(10):
    if(new_df.loc[i, '분실 장소'] == '영일대 설머리물회지구 버스 정류장, 302번 버스, 한동대학교'):
        new_df.loc[i, '분실 장소'] = '버스' 
    elif(new_df.loc[i, '분실 장소'] == '302번 버스 (한동대-&gt; 문덕차고지)'):
          new_df.loc[i, '분실 장소'] = '버스'
    else:
         new_df.loc[i, '분실 장소'] = '한동대학교'
```
