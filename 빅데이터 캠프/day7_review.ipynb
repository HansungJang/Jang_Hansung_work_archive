```python
#음원 추출 예제를 활용하여 (함수, 복수페이지)
import requests
import pandas as pd
from bs4 import BeautifulSoup
import json 
import time

#(가수이름, 노래제목) 동일 데이타에 대해서 출력하는 함수
def make_list(url):
    page = requests.get(url)
    music = json.loads(page.text)
    title = [x['name'] for x in music['data']['list']]
    artist = [x['artistList'][0]['name'] for x in music['data']['list']]
    return pd.DataFrame({'제목': title, 
                  '가수': artist})
    

for i in range(1,3):
    url = 'https://www.music-flo.com/api/meta/v1/track/KPOP/new?page={}&size=100%D7tamp'.format(i)
    df = make_list(url)
    df #정상작동 하지만 보이지 않음
    time.sleep(10)

#solution 
#1. print(df)
#2. df를 for loop 밖에서 출력 
```
