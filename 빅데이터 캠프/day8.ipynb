```python
#서브 플랏3
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400) #numpy에서 간격생성 : ((시작, 끝), 간격)
y = np.sin(x ** 2)

fig, ax = plt.subplots() #graph 1가지만 있는 경우 
ax.plot(x, y)
#ax.set_title('A single plot')
```

```python
fig, axs = plt.subplots(2) #graph 2개 들어간 경우 default: row방향
fig.suptitle('Vertically stacked subplots')
axs[0].plot(x,y)
axs[1].plot(x,-y)
plt.show()
```

```python
#그래프 4개를 그리는 경우 (row X col : 사용가능)
fig, axs = plt.subplots(2,2)
axs[0,0].plot(x,y)
axs[0,0].set_title('Axis [0,0]')
axs[0,1].plot(x,y, 'tab:orange')
axs[0,1].set_title('Axis [0,1]')
axs[1,0].plot(x,-y, 'tab:green')
axs[1,0].set_title('Axis [1,0]')
axs[1,1].plot(x,y, 'tab:red')
axs[1,1].set_title('Axis [1,1]')
plt.tight_layout
```

```python
#occupation example
import random
user = pd.read_csv('occupation.txt', delimiter='|')
#user.set_index()
user.set_index('user_id', inplace=True) #colum을 인덱스로 
user.shape

#user['height'] = random.sample(range(110,200), user.shape[0]) #크기 차이로 sample을 사용할수는 없다?
user['height'] = [random.randint(110,200) for i in range(943)]
user['weight'] = [random.randint(30,100) for i in range(943)]
user

#3개의 그래프를 사용한 방법 scatter-plot 
plt.subplot(131)
plt.scatter(user.height, user.weight)

plt.subplot(132)
plt.scatter(user.age, user.weight)

plt.subplot(133)
plt.scatter(user.age, user.height)

plt.show()
```

```python
#histogram 
plt.hist(user.age, bins=np.arange(0,101,5))
#plt.tight_layout()
plt.show()
```
